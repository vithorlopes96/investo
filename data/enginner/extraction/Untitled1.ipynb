```python
stocks = ['CVX', 'NVDA', 'AAPL','TM','V','JPM']
```

```python
import pandas as pd
```

```python
df_fin = pd. DataFrame() 
```

```python
import yfinance as yf
for i in stocks:
    data_temp = yf.download(i, period="10y")
    data_temp['stock'] = i
    df_fin  = pd.concat([df_fin,data_temp])
```
