```python
def list_withespaces_trim (list_input):
        return [values.strip() for values in list_input]
```

```python
meta_data = data['Meta Data']
header = meta_data['1. Information']
header = re.sub('and',',',header)
header = re.sub('Intraday','Time,',header)
header = re.sub('close prices','close',header)
header = re.sub(r"\(5min\)",'',header).split(',')
header_fin = list_withespaces_trim(header)

```

```python
def get_keys(dictionary):
    result = []
    for key, value in dictionary.items():
        if type(value) is dict:
            new_keys = get_keys(value)
            result.append(key)
           
        else:
            result.append(key)
    return result
```
